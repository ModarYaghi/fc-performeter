```python
mhpss = hsh_report324[hsh_report324['service'].isin(['int', 're', 'gc', 'ic', 'psfua3', 'psfua6', 'psfua12' ])]
unique_mhpss = mhpss.drop_duplicates(subset=['fcid'])
# unique_mhpss
```

```python
pt = hsh_report324[hsh_report324['service'].isin(['ptint', 'gpt', 'ipt', 'ptfua3', 'ptfua6', 'ptfua12' ])]
unique_pt = pt.drop_duplicates(subset=['fcid'])
# unique_pt
```

```python
td = hsh_report324[hsh_report324['service'].isin(['td', 'cdw', 'adv'])]
unique_td = td.drop_duplicates(subset=['fcid'])
# unique_td
```

```python
trw = hsh_report324[hsh_report324['service'].isin(['trw'])]
unique_trw = trw.drop_duplicates(subset=['fcid'])
unique_trw
```

```python
pei = hsh_report324[hsh_report324['service'].isin(['pei'])]
unique_pei = pei.drop_duplicates(subset=['fcid'])
# unique_pei
```

```python
ref = hsh_report324[hsh_report324['service'].isin(['ref'])]
unique_ref = ref.drop_duplicates(subset=['fcid'])
# unique_ref
```

```python
services = [unique_mhpss, unique_pt, unique_td, unique_trw, unique_pei, unique_ref]
hsh_all_beneficiaries_324 = pd.concat(services)
```
